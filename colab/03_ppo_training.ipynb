# PPO (Proximal Policy Optimization) - FFAI Assistant

Actor-Critic con PPO para policy gradient estable.

```
import tensorflow as tf
from tensorflow import keras
import numpy as np
print(f"TensorFlow: {tf.__version__}")
print(f"GPU: {tf.config.list_physical_devices('GPU')}")
```

```
STATE_SIZE, ACTION_SIZE = 256, 15

# Actor Network (Policy)
actor = keras.Sequential([
    keras.layers.Dense(256, activation='relu', input_shape=(STATE_SIZE,)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(ACTION_SIZE, activation='softmax')
], name='PPO_Actor')

# Critic Network (Value)
critic = keras.Sequential([
    keras.layers.Dense(256, activation='relu', input_shape=(STATE_SIZE,)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(1)
], name='PPO_Critic')

print(f"✓ Actor: {actor.count_params():,} params")
print(f"✓ Critic: {critic.count_params():,} params")
```

```
# Exportar Actor
converter = tf.lite.TFLiteConverter.from_keras_model(actor)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
with open('ppo_actor.tflite', 'wb') as f:
    f.write(converter.convert())

# Exportar Critic
converter2 = tf.lite.TFLiteConverter.from_keras_model(critic)
converter2.optimizations = [tf.lite.Optimize.DEFAULT]
with open('ppo_critic.tflite', 'wb') as f:
    f.write(converter2.convert())

print("✓ Exportados: ppo_actor.tflite, ppo_critic.tflite")
```
