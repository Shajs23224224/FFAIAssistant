# SAC (Soft Actor-Critic) - FFAI Assistant

Actor + Twin Q-Networks con temperature automática.

```
import tensorflow as tf
from tensorflow import keras
import numpy as np
print(f"TensorFlow: {tf.__version__}")
```

```
STATE_SIZE, ACTION_SIZE = 256, 15

# Actor (Stochastic Policy)
actor = keras.Sequential([
    keras.layers.Dense(256, activation='relu', input_shape=(STATE_SIZE,)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(ACTION_SIZE, activation='softmax')
], name='SAC_Actor')

# Twin Q-Networks
def create_q_network():
    return keras.Sequential([
        keras.layers.Dense(256, activation='relu', input_shape=(STATE_SIZE + 1,)),
        keras.layers.Dense(128, activation='relu'),
        keras.layers.Dense(1)
    ])

q1_net = create_q_network()
q2_net = create_q_network()

print(f"✓ Actor: {actor.count_params():,} params")
print(f"✓ Q1: {q1_net.count_params():,} params")
print(f"✓ Q2: {q2_net.count_params():,} params")
```

```
# Exportar modelos
for name, model in [('sac_actor', actor), ('sac_q1', q1_net), ('sac_q2', q2_net)]:
    converter = tf.lite.TFLiteConverter.from_keras_model(model)
    converter.optimizations = [tf.lite.Optimize.DEFAULT]
    with open(f'{name}.tflite', 'wb') as f:
        f.write(converter.convert())
    print(f"✓ Exportado: {name}.tflite")
```
